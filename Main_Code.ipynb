## Create Segments

```python
# Time periods
df['Period'] = pd.cut(df['Date'], bins=3, labels=['P1', 'P2', 'P3'])

# Segments
df['Age_Group'] = pd.cut(df['Consumer_Age'], bins=[0, 26, 42, 58, 100], 
                         labels=['Gen-Z', 'Millennial', 'Gen-X', 'Boomer+'])

df['Health_Segment'] = df['Product_Name'].apply(
    lambda x: 'Healthy' if 'Diet' in x or 'Zest-Bio' in x else 'Normal'
)

df['Price_Segment'] = pd.cut(df['Price_Per_Unit'], bins=[0, 1.25, 2.0, 5], 
                              labels=['Value', 'Mid', 'Premium'], include_lowest=True)

df['Loyalty_Group'] = pd.cut(df['Loyalty_Score'], bins=[0, 40, 70, 100], 
                              labels=['Low', 'Medium', 'High'])

# Brand grouping
df['Brand'] = df['Product_Name'].apply(lambda x: 'Effervesce' if 'Effervesce' in x else x)
```

```python
def categorize_value_seeker(row):
    if row['Price_Per_Unit'] < 1.5:
        return 'Value Seeker'
    elif row['Product_Name'] == 'Private Label':
        return 'Value Seeker'
    elif row['Promo_Type'] in ['BOGO', 'Retailer_App']:
        return 'Value Seeker'
    else:
        return 'Premium Seeker'
        
df['Value_Seeker'] = df.apply(categorize_value_seeker, axis=1)
```

## Market Share by Segment

```python
# Calculate share for a segment
def get_share(segment_col):
    share = df.groupby([segment_col, 'Period', 'Product_Name'])['Units_Sold'].sum()
    share = share / df.groupby([segment_col, 'Period'])['Units_Sold'].sum()
    return (share * 100).unstack(fill_value=0)
```

# Customer Churn from Effervesce to Zest-Bio

```python
p1 = df[df['Period'] == 'P1']
p3 = df[df['Period'] == 'P3']

# Calculate brand share changes
def calculate_churn(segment_col):
    results = []
    
    for segment in df[segment_col].dropna().unique():
        # P1 shares
        p1_seg = p1[p1[segment_col] == segment]
        p1_effer = p1_seg[p1_seg['Brand'] == 'Effervesce']['Units_Sold'].sum() / p1_seg['Units_Sold'].sum() * 100
        p1_zest = p1_seg[p1_seg['Product_Name'] == 'Zest-Bio']['Units_Sold'].sum() / p1_seg['Units_Sold'].sum() * 100
        
        # P3 shares
        p3_seg = p3[p3[segment_col] == segment]
        p3_effer = p3_seg[p3_seg['Brand'] == 'Effervesce']['Units_Sold'].sum() / p3_seg['Units_Sold'].sum() * 100
        p3_zest = p3_seg[p3_seg['Product_Name'] == 'Zest-Bio']['Units_Sold'].sum() / p3_seg['Units_Sold'].sum() * 100
        
        results.append({
            'Segment': segment,
            'Effer_P1': p1_effer,
            'Effer_P3': p3_effer,
            'Effer_Δ': p3_effer - p1_effer,
            'Zest_P1': p1_zest,
            'Zest_P3': p3_zest,
            'Zest_Δ': p3_zest - p1_zest
        })
    
    return pd.DataFrame(results).sort_values('Zest_Δ', ascending=False)

#for each segment type
health_churn = calculate_churn('Health_Segment')
age_churn = calculate_churn('Age_Group')
price_churn = calculate_churn('Price_Segment')
loyalty_churn = calculate_churn('Loyalty_Group')
value_seeker = calculate_churn('Value_Seeker')
```

## Gen-Z Pack Size Misalignment Analysis

### Key Finding: Gen-Z Under-Adopts Large Packs Despite Buying Premium Products

**Pack Size Preference by Age Group:**
- **Gen-Z**: 8.7% Large | 91.3% Regular
- **Millennials**: 59.4% Large | 40.6% Regular

- **Market Average**: 40.6% Large packs

**Gen-Z Index: 21** (8.7% ÷ 40.6% × 100) = **79% BELOW market average for large packs**


### Gen-Z Pack Economics Reveal the Problem

**Large Pack Purchases (Gen-Z):**
- Effervesce Classic: $1.99, 165 units
- Effervesce Diet: $1.99, 163 units
- **Zest-Bio: $3.49, 187 units** (highest volume!)

**Regular Pack Purchases (Gen-Z):**
- Sparkle-H2O: $1.25, 5,239 units (dominates)
- Private Label: $1.25, 173 units


### What This Means

**1. Price Ladder Misalignment:**
Gen-Z wants to buy premium products (Zest-Bio at $3.49) but only in LARGE packs, avoiding regular packs almost entirely. Yet they're massively under-indexed on large packs overall (8.7% vs 40.6% market average).

**2. Commitment Barrier:**
- Large packs require upfront investment + commitment to one product
- Gen-Z is trying to buy premium (Zest-Bio large packs = 187 units)
- But overall large pack adoption is LOW because they can't commit or afford bulk premium consistently

**3. Regular Pack Trap:**
When Gen-Z buys regular packs (91.3% of their purchases), they're buying the CHEAPEST options (Sparkle-H2O $1.25 dominates with 5,239 units). They're not buying regular packs of premium products.

**4. Missing Middle:**
There's no "trial-size premium" option. Gen-Z wants to experiment with Zest-Bio but can't afford/commit to large packs consistently, so they default to cheap regular packs (Sparkle-H2O) for everyday consumption.



### Strategic Recommendation

**Current Price Ladder is Failing Gen-Z:**

**Large Premium Packs (3.49)**: Too expensive/risky for consistent Gen-Z purchase
**Regular Value Packs (1.25)**: Too cheap to build premium brand affinity

**SOLUTION: Introduce "Gen-Z Medium" Pack Size**
- **Target Price**: 2.49 (between 1.99 Effervesce and 3.49 Zest-Bio)
- **Size**: Smaller than large, bigger than regular (6-8 units vs 4 regular/12 large)
- **Positioning**: "Trial premium" or "on-the-go premium"
- **Goal**: Capture Gen-Z when they want premium but can't commit to bulk

This bridges the gap between their value purchases (91.3% regular packs at 1.25) and their aspiration for premium (Zest-Bio large at 3.49), allowing more frequent premium engagement without the commitment barrier.
